# Exploratory Data Analysis (Selected Dataset Only)

For EDA, I will be focusing on one day of the CIC-IDS-2018. The full dataset contains more days with similar structure.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv(
    "/content/Friday-02-03-2018_TrafficForML_CICFlowMeter.csv",
    low_memory=False,

)

assert df.shape[0] > 0 and df.shape[1] > 0, "dataFrame is empty after loading"
assert "Label" in df.columns, "expected 'Label' column not found"

df["Label"] = df["Label"].astype(str).str.strip()
df.loc[df["Label"].isin(["", "nan", "NaN", "None"]), "Label"] = np.nan
missing_labels = df["Label"].isna().sum()
print("missing labels before drop: ", missing_labels)

df = df.dropna(subset=["Label"])


#observing column type
display(df.sample(5))
df.info()

#diplaying missing entries
missing = df.isna().sum()
display(missing)

#table distribution
display(df['Label'].value_counts())
display(df['Label'].value_counts(normalize=True))

df['Label'].value_counts().plot(kind='bar', figsize=(8,4))
plt.ylabel('count')
plt.title('label distribution')
plt.show()
```

```python
#convert timestamp to datetime format
df.columns = [c.strip() for c in df.columns]
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')

print("Timestamp nulls:", df['Timestamp'].isna().sum())

#resample traffic into minute intervals
df_time = df.dropna(subset=['Timestamp']).set_index('Timestamp')
flows_per_min = df_time.resample('1min').size()

plt.figure(figsize=(12,4))
flows_per_min.plot()
plt.title("flows per minute")
plt.ylabel("flow count")
plt.tight_layout()
plt.show()
```

```python
#plot table to show whether traffic spices correlate to attack activity or normal behavior
flows_per_min_label = (
    df_time.groupby("Label")
           .resample("1min")
           .size()
           .unstack(0)
           .fillna(0)
)

plt.figure(figsize=(12,4))
flows_per_min_label.plot(figsize=(12,4))
plt.title("flows per minute by label")
plt.ylabel("flow count")
plt.tight_layout()
plt.show()
```

```python
#analyze feature distributions by trimming extreme values to improve readability
features = ["Flow Duration", "Tot Fwd Pkts", "Tot Bwd Pkts", "TotLen Fwd Pkts", "TotLen Bwd Pkts"]

for f in features:
    if f in df.columns:

      #removing infinite values and outiers
        x = df[f].replace([np.inf, -np.inf], np.nan).dropna()
        x = x.clip(lower=x.quantile(0.01), upper=x.quantile(0.99))
        plt.figure(figsize=(6,4))
        plt.hist(x, bins=50)
        plt.title(f"distribution of {f} (1%-99% clipped)")
        plt.ylabel("count")
        plt.tight_layout()
        plt.show()
```

```python
#sample subset of data to compute the correlations
sample_df = df.sample(n=min(200_000, len(df)), random_state=42)
corr_cols = sample_df.select_dtypes(include=[np.number]).columns[:25]
corr = sample_df[corr_cols].corr()

plt.figure(figsize=(10,8))
plt.imshow(corr, aspect="auto")
plt.colorbar()
plt.xticks(range(len(corr_cols)), corr_cols, rotation=90)
plt.yticks(range(len(corr_cols)), corr_cols)
plt.title("correlation (subset of numeric features)")
plt.tight_layout()
plt.show()

```

During EDA, I made some decistions to improve the quality of the data, interpretability, and feasibility.

First, the timestamp column was converter to datetime format. Invalid timestamp entries were coerced to NaT to prevent runtime errors during time-based operations. Rows with invalid timestamps were dropped before resampling. This was necessary because time-series operations like resampling need a datetime index. Using invalid timestamps would lead to misleading traffic patterns.

Network flows were aggregated into 1 minute intervals for two reasons. It was implemented for noise reduction because raw data can differe pretty drastically. Aggregating to 1 minute intervals can smooth out short term noise and make larger traffic patterns more visible. Along with this, many network attacks happen in a short time so 1 minute intervals allow for balance between temporal resolution and interpretability.

I replaced infinite values and between 1%-99% range to better visualize the data. The CIC-IDS dataset is extremely skewed. A small number of flows dominate the range and without clipping, the histograms would be unreadable due to extreme outliers.

A random subset of 200,000 rows was used to compute feature correlations. The full dataset has millions of flows, there amount of features would be too large, and plotting all the correlations would increase runtime and memory usage drastically. Using a large random sample preserves the representation structure while reducing memory usage and runtime.

Finally, I used the first 25 numeric features to create a heatmap. This was done to show what is most used. The dataset contains multiple numeric features, and plotting that would be unreadable. The smaller subset identifies redundancy patterns while still capturing the structure.
